```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import metrics
```

```python
names=['sepal-length','sepal-width','petal-length','petal-width','Class']
dataset=pd.read_csv('../data/iris.data',names=names)
dataset
```

```python
X=dataset.values[:,:-1]
Y=dataset.values[:,-1]
#분류
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
X_train[0:10]
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
X_train=s.fit_transform(X_train)
X_test=s.fit_transform(X_test)
print(X_train[0:5])
print(X_test[0:5])
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=50)
knn.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=knn.predict(X_test)
print('정확도:{}'.format(accuracy_score(y_test,y_pred)))
```

```python
print(y_pred[0:10])
print(y_test[0:10])
```

```python
k=50

acc_array=np.zeros(k)
for i in np.arange(1,k+1,1):
  classifier=KNeighborsClassifier(n_neighbors=i).fit(X_train,y_train)
  y_pred=classifier.predict(X_test)
  acc=metrics.accuracy_score(y_test,y_pred)
  acc_array[i-1]=acc
  
max_acc=np.amax(acc_array)
acc_list=list(acc_array)
k=acc_list.index(max_acc)
print(acc_list)
print('정확도 {}으로 최적의 K는 {}입니다'.format(max_acc,k+1))

```

```python
from sklearn import svm
from sklearn import metrics
from sklearn import datasets
from sklearn import model_selection

import tensorflow as tf

import os
os.environ['TF_CPP_MIN_LOG_LEVEL']='3'
```

```python
iris=datasets.load_iris()
iris
```

```python
x_train,x_test,y_train,y_test=model_selection.train_test_split(iris.data,iris.target,test_size=0.5,random_state=33)
```

```python
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
svm=svm.SVC(kernel='linear',C=1.0, gamma=0.5)
svm.fit(x_train, y_train)
prediction=svm.predict(x_test)
score=metrics.accuracy_score(y_test, prediction)
print('정확도:{0:f}'.format(score))
```

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/titanic_train.csv', index_col='PassengerId')
print(df.head())
```

```python
df=df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']]
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1}) # 성별을 0과 1로 변환
df = df.dropna() # 결측치가 있는 행을 제거
X = df.drop('Survived', axis=1) # 'Survived' 열을 제외한 모든 열
y = df['Survived'] # 'Survived' 열
y # dataframe에서 하나의 값은 series로 반환
```

```python
df.info()
```

```python
# train, test 데이터 나누기

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier() # 의사결정트리 모델을 사용
model.fit(X_train,y_train) 

# 예측 결과 출력
```

```python
y_predict = model.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict) # 정확도를 출력
```

```python
from sklearn.metrics import confusion_matrix
pd.DataFrame(
    confusion_matrix(y_test, y_predict),
    columns=['Predicted Not Survival', 'Predicted Survival'],
    index=['True Not Survival', 'True Survival']
)
```
